```python
#3
 
legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}
    
def eta (legs, source, destination):
    
    time_length = 0
    
    start = [route[0] for route in legs]
    end = [route[1] for route in legs]
    
    time_length = time_length + (list(legs.values())[start.index(source)]["travel_time_mins"])
    
    while True: 
        if destination == end[start.index(source)]:
            break
        
        source = end[start.index(source)]
        time_length = time_length + (list(legs.values())[start.index(source)]["travel_time_mins"])
        
    return time_length

eta(legs, "upd", "dlsu")
```
